```python
import matplotlib.pyplot as plt
categories=['transport','income tax','housing']
values1=[15,24,30]
values2=[20,18,25]

#creating the first set of bars(group1) without any offset
plt.bar(categories,values1,label='first half',color='red')
#creating the second set of bars(group2) plotted with 'bottom' set to the values
#this makes group 2 bars stacked on top of group 1 bar
plt.bar(categories,values2,bottom=values1,label='second half',color='pink')

#adding labels to the axes
plt.xlabel('categories')
plt.ylabel('values')
plt.title('stacked bar graph')
plt.legend()
plt.show()
```

```python
import pandas as pd
#series generated from a list
a=[1,7,2]
myvar=pd.Series(a)
print(a)
print(myvar)
print('-------------')
#labels
print(myvar[0])
print('-------------')
#assigning seperate labels to values in the series
myvar2=pd.Series(a,index=["x","y","z"])
print(myvar2)
print(myvar2["y"])
print('-------------')
#series generated from dictionary
calories={"day1":420,"day2":380,"day3":390}
myvar=pd.Series(calories)
print(myvar)
print('-------------')
myvar=pd.Series(calories,index=["day1","day2"])
print(myvar)
print('-------------')
#series generated from scalar value
series=pd.Series(5)
print(series)
index=[0,1,2,3]
print('-------------')

```

```python
#integer series
int_series=pd.Series([1,2,3,4])
print(int_series)
print('-------------')
#float series
float_series=pd.Series([1.1,2.2,3.3,4.4])
print(float_series)
print('-------------')
#string series
string_series=pd.Series(['a','b','c','d'])
print(string_series)
print('-------------')
#label-based indexing
data={'a':10,'b':20,'c':30}
series=pd.Series(data)
print(series['a'])
print(series[['a','c']])
print('-------------')
#boolean indexing
data=[10,20,30,40,50]
series=pd.Series(data)
#access elements greater than 25
print(series[series>25])
print('-------------')

```

```python
#creating a dataframe
data={
    'Name':['Alice','Bob','Charlie'],
    'Age':[25,30,35],
    'City':['New York','Los Angeles','Chicago']
}
df=pd.DataFrame(data)
print('-------------')
#selecting colums
print(df.Name)
print(df['Name'])
print(df[['Name','City']])
print('-------------')
#select a single row by label
print(df.loc[0])
print(df.iloc[0])
#select multiple rows
print(df.iloc[0:2])
print('-------------')
#conditional selection
data={
    'Name':['alice','Bob','Charlie'],
    'Age':['24,19,40'],
    'City':['New York','Los Angeles','Chicago']
}
df=pd.DataFrame(data)
#select rows where age is greater than 25
print(df[df['Age']>25])
print('-------------')
print(df.loc[(df.Name=='Bob')&(df.City=='Los Angeles')])

```
